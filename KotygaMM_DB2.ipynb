`
Вывести все подтвержденные транзакции за период '2017-04-01' по '2017-04-09' включительно.
`

```sql
select *
from sales.transaction_kotyga
where order_status = 'Approved'
and transaction_date::date between '2017-04-01'::date and '2017-04-09'::date
```

# Задание 3 (1 балл)

`
Вывести все профессии у клиентов из сферы IT или Financial Services, которые начинаются с фразы 'Senior'.
`

```sql
select distinct job_title
from sales.customer_kotyga
where job_industry_category = any(array['IT', 'Financial Services'])
and job_title like 'Senior%'
```

# Задание 4 (1 балл)

`
Вывести все бренды, которые закупают клиенты, работающие в сфере Financial Services.
`

```sql
select distinct brand
from sales.transaction_kotyga tk
join sales.customer_kotyga ck on ck.customer_id = tk.customer_id
where ck.job_industry_category = 'Financial Services'
```

# Задание 5 (1 балл)

`
Вывести 10 клиентов, которые оформили онлайн-заказ продукции из брендов 'Giant Bicycles', 'Norco Bicycles', 'Trek Bicycles'.
`

```sql
select distinct ck.customer_id, first_name, last_name, gender, dob
from sales.customer_kotyga ck
join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
where tk.brand = any(array['Giant Bicycles', 'Norco Bicycles', 'Trek Bicycles'])
limit 10
```

`
Вывести всех клиентов из IT, у которых транзакции с максимальной стандартной стоимостью.
`

```sql
select distinct ck.customer_id, first_name, last_name, gender, dob
from sales.customer_kotyga ck
left join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
where job_industry_category = any(array['IT'])
and standard_cost = (select max(standard_cost) from sales.transaction_kotyga)
```

# Задание 8 (2 балла)

`
Вывести всех клиентов из сферы IT и Health, у которых есть подтвержденные транзакции за период '2017-07-07' по '2017-07-17'.
`

```sql
select distinct ck.customer_id, first_name, last_name, gender, dob
from sales.customer_kotyga ck
left join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
where job_industry_category = any(array['IT', 'Health'])
and order_status = 'Approved'
and transaction_date::date between '2017-07-07'::date and '2017-07-17'::date
```
