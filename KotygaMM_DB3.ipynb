# Задание 2 (1 балл)

`
Найти сумму транзакций за каждый месяц по сферам деятельности, отсортировав по месяцам и по сфере деятельности.
`

```sql
with time_series as (
	SELECT generate_series(
    (SELECT min(transaction_date::date) FROM sales.transaction_kotyga),
    (SELECT max(transaction_date::date) FROM sales.transaction_kotyga),
    interval '1 month'
	)::date ts
) -- чтобы быть уверенным, что у нас точно есть такая дата, даже если нет транзакций

select
  ts,
  job_industry_category,
  count(distinct transaction_id) trans_count,
  sum(standard_cost) sum_trans
from time_series
left join sales.transaction_kotyga tk on time_series.ts = tk.transaction_date::date
left join sales.customer_kotyga ck on ck.customer_id = tk.customer_id
group by 1, 2
order by 1, 2
```

# Задание 3 (1 балл)

`
Вывести количество онлайн-заказов для всех брендов в рамках подтвержденных заказов клиентов из сферы IT.
`

```sql
select
	tk.brand,
	count(distinct transaction_id) filter (where online_order = 'True')

from sales.transaction_kotyga tk
left join sales.customer_kotyga ck on ck.customer_id = tk.customer_id

where order_status = 'Approved' and
	  ck.job_industry_category = 'IT'
	  
group by 1
```

# Задание 5 (2 балл)

`
Найти имена и фамилии клиентов с минимальной/максимальной суммой транзакций за весь период (сумма транзакций не может быть null). Напишите отдельные запросы для минимальной и максимальной суммы.
`

```sql
--Макс
select
    ck.first_name,
    ck.last_name,
    sum(tk.list_price) total_list_price
from sales.customer_kotyga ck
left join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
group by ck.first_name, ck.last_name
having sum(tk.list_price) = (
    select max(total)
    from (
        select sum(list_price) total
        from sales.transaction_kotyga
        group by customer_id
    ) sub
)


-- ИЛИ
select
    ck.first_name,
    ck.last_name,
    sum(tk.list_price) total_list_price
from sales.customer_kotyga ck
left join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
where list_price is not null
group by ck.first_name, ck.last_name
order by 3 desc
limit 1


-- Мин
select
    ck.first_name,
    ck.last_name,
    sum(tk.list_price) total_list_price
from sales.customer_kotyga ck
left join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
group by ck.first_name, ck.last_name
having sum(tk.list_price) = (
    select min(total)
    from (
        select sum(list_price) total
        from sales.transaction_kotyga
        group by customer_id
    ) sub
)


-- ИЛИ
select
    ck.first_name,
    ck.last_name,
    sum(tk.list_price) total_list_price
from sales.customer_kotyga ck
left join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
where list_price is not null
group by ck.first_name, ck.last_name
order by 3
limit 1
```

# Задание 6 (1 балл)

`
Вывести только самые первые транзакции клиентов. Решить с помощью оконных функций.
`

```sql
with first_trans as (
select distinct
transaction_id,
customer_id,
transaction_date::date,
row_number() over (partition by customer_id order by transaction_date::date) p
from sales.transaction_kotyga tk
)

select
transaction_id,
customer_id,
transaction_date
from first_trans
where p = 1
```

# Задание 7 (2 балла)

`
Вывести имена, фамилии и профессии клиентов, между транзакциями которых был максимальный интервал (интервал вычисляется в днях)
`

```sql
with transaction_intervals AS (
    select
    	ck.customer_id,
        ck.first_name,
        ck.last_name,
        ck.job_title profession,
        tk.transaction_date,
        lag(tk.transaction_date) over (partition by ck.customer_id order by tk.transaction_date)::date previous_transaction_date
    from sales.customer_kotyga ck
    join sales.transaction_kotyga tk on ck.customer_id = tk.customer_id
),
intervals as (
    select
        customer_id,
        first_name,
        last_name,
        profession,
        transaction_date::date,
        previous_transaction_date,
        (transaction_date::date - previous_transaction_date) interval_days
    from transaction_intervals
    where previous_transaction_date is not null
)
select
    customer_id,
    first_name,
    last_name,
    profession
from intervals
where interval_days = (
    select max(interval_days)
    from intervals
)
```
